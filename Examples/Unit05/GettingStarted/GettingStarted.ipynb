## 5. loc and iloc
Methods for accessing part of the DataFrame

### iloc
Uses the number values for the indices of rows and columns

```python
# Get the element in the first row, 3rd column
df.iloc[0,2]

# Get the first 5 rows and first 5 columns
df.iloc[:5, :5]

# Get every row, but only second and third column
df.iloc[:, 1:3] # or [:, [1,2]]

# Will give 6th row (AS SERIES)
df.iloc[6]

# Will give 6th row (AS DATAFRAME)
df.iloc[[5]]
df.iloc[5:6]
```

### loc
Uses index names and column names

```python
# Get every row but only certain values
df.loc[:, ["BEDS", "BATH"]]

# All rows with more than 5 bedrooms
df.loc[df["BEDS"] > 5]

# Filter on multiple conditions (&, |, ~)
df.loc[(df["BEDS"] > 5) & (df["BEDS" == 2])]
```

## 6. Querying Data

```python
# Basic Query (query is slower than loc)
df.query("(BEDS > 5) and (BATH == 2)")

# Using strings in the query by wrapping them in single quotes
df.query("(BEDS > 5) and (SUBLOCALITY == 'Bronx County')")

# Checking if sublocality contains Bronx
df.query("SUBLOCALITY.str.contains('Bronx')")

# using variables in our boolean expressions
county = "Bronx County"
df.query("(BEDS > 5) and (SUBLOCALITY == @county)")
```
